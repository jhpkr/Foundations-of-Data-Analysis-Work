1.The UVA men’s and women’s basketball teams are both playing Virginia Tech on the same
day. There is a 60% chance that the UVA men win and a 60% chance that the UVA women
win. In addition, there is a 75% chance that UVA wins at least one of the games (that is,
they win the men’s game, or the women’s game, or both). Let M denote the event that the
UVA men win, and let W denote the event that the UVA women win.
- What does P(M ∪ W) mean in English? What is its value?
- What is the probability that UVA wins both games?
- What does P(M | W) mean in English? What is its value?
- What is the probability that UVA wins exactly one game (men or women, but not both)?
- If the UVA women lose, what is the probability that the UVA men also lose? 

1a. P(M U W) represents the probability that either men or women OR both men or women win. Given in the question, it says that there is a 75% chance that UVA men, women or both teams win. Therefore it is 75%


1b. Probability that UVA wins both games; using the inclusion-exclusion  = 
$$P(M \cup W) = P(M) + P(W) − P(M ∩ W)$$

$$.75 = .6 + .6 - P(M ∩ W)$$

$$P(M ∩ W) = .45$$


1c. P(M | W) represents the probability of men winning the game given that we know that women won.
$$P(M|W) = P(M \cap W)/P(W)$$
$$P(M|W) = .45/0.6$$
$$P(M|W) = .75$$

1d. If UVA wins exactly 1 game we must subtract the probability that UVA wins both games from P(M U W):
$$= P(M \cup W) - P(M \cap W)$$ 
$$= .75 - .45 = .30$$

1e. If UVA women lose, the probability that UVA men also lose is: 
$$(P(M^c \cap W^c))/P(W^c)$$
$$P(M^c \cap W^c) = 1 - P(M \cup W)$$
$$ (1 - .75 / 0.4) = .625$$

2. A man is accused of murder, and the prosecutor has fingerprint evidence from the crime scene
that matches the defendant. There is only a 1% chance of a fingerprint match if the defendant
were not guilty. However, there are 101 people in the town where the murder was committed,
that is, there is a 1/101 chance that the man is guilty when the evidence is not taken into
account. What is the probability that the defendant is guilty, given the fingerprint evidence?

Notes: P(G) = Probability that they are guilty; P(M) = Probability that the fingerprint matches; $$P(M|G^c) = .01$$ $$P(G) = 1/101$$

Using Bayes Rule, we are able to find the probability of the defendant being guilty given the fingerprint evidence:
$$ P(G|M) = (P(M|G)*P(G))/P(M))$$
$$ = (1 * (1/101))/P(M)$$ 

Note: For this problem, probability of someone's finger print matching given they are guilty is 1

## Finding P(M):
Using total probability rule found on Prof. Fletcher's Probability Cheat sheet:
$$P(M) = P(M|G)P(G) + P(M|G^c)P(G^c)$$
$$P(M) = (1 * 1/101) + (0.01*100/101) = 0.0198...$$

## Solution:
$$(1/101)/(0.0198...) = .5$$


3. According to the American Lung Association, there is a 0.13% chance to develop lung cancer.
Of the people who have lung cancer, 90% of them are smokers. In the population of people
who do not have lung cancer, 16.9% are smokers.
- What percentage of the total population are smokers?
- If you are a smoker, what is your probability to develop lung cancer?
- If you are not a smoker, what is your probability to develop lung cancer?

$$P(L)*P(S|L) + P(S|L^c)*P(L^c)$$
$$0.0013 * 0.9 + (1-0.0013)*0.169 = .1699503$$ 

b.Probability to develop lung cancer:
$$P(L|S) = (P(S|L)P(L))/P(S)$$
$$= (0.9 * 0.0013)/.1699503 = 0.0069$$ 
Note: .1699503 is found from the answer above

c.Probability to develop lung cancer if you are not a smoker:
$$P(L|S^c) = (P(S^c|L)P(L))/(P(S^c)$$
$$P(S^c|L) = 1- P(S|L) = 0.1$$
$$(0.1 * 0.0013)/(P(S^c))$$
$$P(S^C) = 1 - 0.1699503 = 0.8300$$

Equations

```python
def gaussian (row, mean, variance):
    denom1 = math.pi * 2
    denom2 = denom1 * variance
    denom3 = math.sqrt(denom2)
    
    first = 1/denom3
    
    num1 = row - mean
    num2 = num1 **2
    num3 = -1 * num2 
    num4 = 2 * variance
    num5 = num3/num4
    
    exp = math.exp(num5)
    gaus = first * exp
    return gaus
```

```python
dfTest = df[df['TrainData'] == 0]
```

```python
dfTestDementia = dfTest[dfTest['Dementia'] == 1]
dfTestNoDementia = dfTest[dfTest['Dementia'] == 0]
```

## Actual Calcuations

```python
dfTest
```

```python
dfTest["right_given_dementia"] = dfTest['RightHippoVol'].apply(gaussian,mean = dementia_mean_right, variance = dementia_variance_right)
dfTest["left_given_dementia"] = dfTest['LeftHippoVol'].apply(gaussian,mean = dementia_mean_left, variance = dementia_variance_left)

dfTest["right_given_no_dementia"] = dfTest['RightHippoVol'].apply(gaussian,mean = no_dementia_mean_right, variance = no_dementia_variance_right)
dfTest["left_given_no_dementia"] = dfTest['LeftHippoVol'].apply(gaussian,mean = no_dementia_mean_left, variance = no_dementia_variance_left)
```

```python
dfTest['evidence'] = (0.5 * dfTest["right_given_dementia"] * dfTest["left_given_dementia"]) + (0.5 * dfTest["right_given_no_dementia"] * dfTest["left_given_no_dementia"])
```

```python
dfTest
```

```python
dfTest['dementia_prediction'] = (0.5 * dfTest["right_given_dementia"] * dfTest["left_given_dementia"])/dfTest['evidence']
dfTest['no_dementia_prediction'] = (0.5 * dfTest["right_given_no_dementia"] * dfTest["left_given_no_dementia"])/dfTest['evidence']
```

```python
dfTest
```

```python
dfTest['Prediction'] = np.where(dfTest['dementia_prediction'] > dfTest['no_dementia_prediction'] ,1, 0)
dfTest
```

```python
dfTest['Accuracy'] = np.where(dfTest['Prediction'] != dfTest['Dementia'] , "Wrong", "Correct")
dfTest
```
